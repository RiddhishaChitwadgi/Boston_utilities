```python
date_columns = ['InvoiceDate', 'FromDate', 'ToDate']
for column in date_columns:
    df[column] = pd.to_datetime(df[column])

df['Year'] = df['InvoiceDate'].dt.year
df
```

```python
# Define the new column names
new_column_names = {
    'InvoiceID': 'Invoice_ID',
    'AccountNumber': 'Account_Number',
    'EnergyTypeName': 'Energy_Type_Name',
    'InvoiceDate': 'Invoice_Date',
    'FromDate': 'From_Date',
    'ToDate': 'To_Date',
    'UsagePeriodDays': 'Usage_Period_Days',
    'DeliveryCost': 'Delivery_Cost',
    'SupplyCost': 'Supply_Cost',
    'TotalCost': 'Total_Cost',
    'TotalConsumption': 'Total_Consumption',
    'DemandkW': 'Demand_kW',
    'UomName': 'Uom_Name',
    'StreetAddress': 'Street_Address',
    'City': 'City',
    'Zip': 'Zip',
    'StateName': 'State_Name',
    'Abbreviation': 'Abbreviation',
    'CountryName': 'Country_Name',
    'SiteName': 'Site_Name',
    'Currency': 'Currency',
    'CodeDescription': 'Code_Description',
    'DepartmentName': 'Department_Name'
}

# Rename the columns
mydata = df.rename(columns=new_column_names)
mydata
```

```python
# Impute "Massachusetts" in the "StateName" column where it is missing
mydata['State_Name'] = mydata['State_Name'].fillna('Massachusetts')
# Impute "MA" in the "StateName" column where it is missing
mydata['Abbreviation'] = mydata['Abbreviation'].fillna('MA')
mydata
```

```python
# Change the name of a categorical variable in the "EnergyTypeName" column
mydata['Energy_Type_Name'] = mydata['Energy_Type_Name'].replace('#2 Oil', 'Oil')
```

```python
# Drop the "UomName" and "DemandkW" columns
mydata = mydata.drop(['Uom_Name', 'Demand_kW'], axis=1)
mydata
```

```python
# Calculate the average total consumption over the years and group by year and energy type
average_consumption = mydata.groupby(['Year', 'Energy_Type_Name'])['Total_Consumption'].mean()

# Unstack the multi-index to create a pivot table
pivot_table = average_consumption.unstack()

pivot_table
```

```python
# Calculate the average total consumption over the years and group by year and energy type
average_consumption = mydata.groupby(['Year', 'Energy_Type_Name'])['Total_Consumption'].mean()

# Convert the multi-index series to a dictionary
average_consumption_dict = average_consumption.to_dict()

# Function for imputation
def impute_total_consumption(row):
    energy_type = row['Energy_Type_Name']
    year = row['Year']
    total_consumption = row['Total_Consumption']
    
    if pd.isnull(total_consumption):
        imputed_value = average_consumption_dict.get((year, energy_type), None)
        return imputed_value
    else:
        return total_consumption

# Apply the imputation function to fill null values in Total_Consumption column
mydata['Total_Consumption'] = mydata.apply(impute_total_consumption, axis=1)
```

```python
# Plot the graph
pivot_table.plot(kind='line', marker='o', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Average Total Consumption')
plt.title('Average Total Consumption over the Years, Grouped by Energy Type')
plt.legend(title='Energy Type')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Count the occurrences of each categorical variable in "DepartmentName" grouped by "EnergyTypeName"
counts = mydata.groupby('Energy_Type_Name')['Department_Name'].value_counts()

# Unstack the multi-index to create a pivot table
pivot_table1 = counts.unstack()

pivot_table1
```

```python
# Calculate the mode for each energy type based on 'Energy_Type_Name'
mode_departments = counts.groupby('Energy_Type_Name').apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None).to_dict()

# Define the imputation function
def impute_department_name(row):
    energy_type = row['Energy_Type_Name']
    department_name = row['Department_Name']
    
    if pd.isnull(department_name):
        imputed_value = mode_departments.get(energy_type, None)
        return imputed_value
    else:
        return department_name

# Apply the imputation function to fill null values in the 'Department_Name' column
mydata['Department_Name'] = mydata.apply(impute_department_name, axis=1)
```
